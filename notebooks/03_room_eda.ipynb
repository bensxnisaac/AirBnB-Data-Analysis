# Room Exploratory Data Analysis

This notebook explores the Room dataset through exploratory data analysis (EDA).

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the processed listings data
df = pd.read_csv("../data/processed/processed_listings.csv")
```

```python
# Room type distribution
room_types = df["room_type"].value_counts()
room_types
```

```python
# Plotting the distribution of room types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="room_type", order=room_types.index, palette="Set2", hue="room_type", legend=False,)
plt.title("Distribution of Room Types", fontsize=16)
plt.xlabel("Room Type", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.savefig("../outputs/room_type_distribution.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# Average price by room type
df.groupby("room_type")["price"].mean().sort_values(ascending=False)
```

```python
# Average availability by room type
df.groupby("room_type")["availability_365"].mean()
```

```python
# Total number of reviews per room type
total_reviews = df.groupby("room_type")["number_of_reviews"].sum()
total_reviews
```

```python
# Review counts by room type
df.groupby("room_type")["number_of_reviews"].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="room_type", y="price", data=df)
plt.title("Price Distribution by Room Type", fontsize=16)
plt.xlabel("Room Type", fontsize=14)
plt.ylabel("Price ($)", fontsize=14)
plt.savefig("../outputs/price_distribution_by_room_type.png", dpi=150, bbox_inches="tight")
plt.show()
```

# Summary of Room Exploratory Data Analysis

This exploratory data analysis (EDA) examines Airbnb listings data from the Room dataset. Key insights include:

- **Room Type Distribution**: Entire home/apt is the most common (25,407 listings), followed by Private room (22,319) and Shared room (1,158).
- **Pricing**: Entire home/apt has the highest average price, with significant variation shown in the boxplot.
- **Availability**: Average availability over 365 days varies by room type, with Entire home/apt showing higher availability.
- **Reviews**: Entire home/apt has the highest total reviews (580,400), followed by Private room (537,979) and Shared room (19,249). Average reviews per listing follow a similar pattern.

Visualizations include count plots for room types and boxplots for price distribution, saved as PNG files.

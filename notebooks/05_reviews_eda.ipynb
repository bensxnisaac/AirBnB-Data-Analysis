# Reviews Exploratory Data Analysis

This notebook explores the relationship between reviews and other listing characteristics.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/processed_listings.csv")
```

```python
# Listing popularity by number of reviews
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, y="number_of_reviews", x="price", alpha=0.8, palette="Set2", hue="room_type", legend=False)
plt.xlabel("Price ($)", fontsize=14)
plt.ylabel("Number of Reviews", fontsize=14)
plt.title("Listing Popularity by Number of Reviews and Price", fontsize=16)
plt.savefig("../outputs/listing_popularity_by_number_of_reviews_and_price.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
df[["price", "number_of_reviews"]].corr()
```

```python
numeric_cols = [
    "price",
    "minimum_nights",
    "number_of_reviews",
    "reviews_per_month",
    "availability_365",
    "calculated_host_listings_count"
]

plt.figure(figsize=(10,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix", fontsize=16)
plt.savefig("../outputs/correlation_matrix.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Summary

**Key Findings:**

- **Price vs Reviews**: There is a weak negative correlation (-0.048) between price and number of reviews, suggesting that cheaper listings tend to receive slightly more reviews.
- **Review Distribution**: The scatter plot shows that most listings with high review counts are concentrated in the lower to mid-price range.
- **Correlation Insights**: The correlation matrix reveals that review-related metrics have minimal correlation with other numeric features, indicating reviews are influenced by factors beyond just price and availability.

## Data Cleaning

This section ensures data quality before analysis.

```python
# Import neccessary libraries
import pandas as pd
```

```python
# Load the dataset into Dataframe
df = pd.read_csv('../data/raw/listings.csv')
```

```python
# Quick inspection
df.head()
```

```python
# Check for duplicated values
df.duplicated().sum()
```

```python
# Check for missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Check for missing values in percentage
(df.isnull().mean() * 100).sort_values(ascending=False)
```

```python
# Fill missing values in "name"  with "Unknown Listing"
df["name"] = df["name"].fillna("Unknown")

# Fill missing values in "host_name" with "Unknown Host"
df["host_name"] = df["host_name"].fillna("Unknown")
```

```python
# Convert last_review to datetime 
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Create a new column "has_review" to indicate if the listing has reviews or not
df['has_review'] = df['number_of_reviews'] > 0
```

```python
# Fill missing values in "reviews_per_month" with 0
df["reviews_per_month"] = pd.to_numeric(df["reviews_per_month"], errors="coerce")

df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
```

```python
# Check for prices that are 0
df[df['price'] == 0]
```

```python
# Remove listings with a price of 0
df = df[df["price"] != 0]
```

```python
# Stripped neighbourhood column
df['neighbourhood'] = df['neighbourhood'].str.strip().str.title()
```

```python
# Remove latitude and longitude from dataset
df = df.drop(columns=["latitude", "longitude"])
```

```python
# Save the cleaned dataset
df.to_csv("../data/processed/processed_listings.csv")
```

### Data Cleaning Summary

- No duplicate rows found.
- Missing host names replaced with "Unknown".
- reviews_per_month missing values filled with 0.
- last_review converted to datetime format.
- Created a new column "has_reviews".
- Listings with prices of 0 are removed.
